FINDING MISSING DATA

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
df
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3)
```

```python
df.fillna(0)
```

```python
df
```

```python
values = {"A":0,"B":100,"C":300,"D":400}
```

```python
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
