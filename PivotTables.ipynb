```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': [
        'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'
    ],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': [
        'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
        'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice'
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
df["Month"] = df["Date"].dt.month_name()
```

```python
df["Quarter"] = "Q" + df["Date"].dt.quarter.astype(str)
```

```python
df
```

```python
pd.pivot_table(df,values="Sales",index = "Region",columns="Product",aggfunc="median")
```

```python
pd.pivot_table(df,values=["Sales","Units"],index = "Region",columns="Product",aggfunc="median")
```

```python
pd.pivot_table(df,values=["Quarter","Units"],index = "Region",columns="Product",aggfunc="median")
```

```python
pd.crosstab(df["Region"],df["Product"])
```
